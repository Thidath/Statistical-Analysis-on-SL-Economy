```python
# prompt: do a hypothisis testing for srilankan gdp

import numpy as np
from scipy import stats

# Load the GDP data for Sri Lanka
df_SriLanka = df[df['Country Name'] == 'Sri Lanka']
gdp_data = df_SriLanka['GDP'].values

# Set the significance level
alpha = 0.05

# Perform a Shapiro-Wilk test for normality
stat, p_value = stats.shapiro(gdp_data)

# Interpret the results
if p_value > alpha:
    print('The GDP data is normally distributed.')
else:
    print('The GDP data is not normally distributed.')

# Perform a one-sample t-test to compare the mean GDP to a hypothesized value
hypothesized_mean = 100000000000  # Replace with the desired hypothesized mean
t_stat, p_value = stats.ttest_1samp(gdp_data, hypothesized_mean)

# Interpret the results
if p_value < alpha:
    print('The mean GDP is significantly different from the hypothesized value.')
else:
    print('The mean GDP is not significantly different from the hypothesized value.')
```

```python
# prompt: Do a ADF teet to check the stationarity of the df_srilanka GDP

from statsmodels.tsa.stattools import adfuller

# Extract the GDP data
gdp_data = df_SriLanka['GDP'].values

# Perform the ADF test
adf_result = adfuller(gdp_data)

# Print the ADF test results
print('ADF Statistic:', adf_result[0])
print('p-value:', adf_result[1])

# Interpret the results
if adf_result[1] > 0.05:
    print('The GDP data is non-stationary.')
else:
    print('The GDP data is stationary.')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from pmdarima.model_selection import train_test_split

# Load the GDP data for Sri Lanka
df_SriLanka = df[df['Country Name'] == 'Sri Lanka']
gdp_data = df_SriLanka['GDP'].values

# Split the data into training and test sets
train, test = train_test_split(gdp_data, train_size=0.8)  # Removed shuffle=False

# Fit an ARIMA model to the training data
model = ARIMA(train, order=(1, 1, 1))
results = model.fit()

# Make predictions on the test set
forecast_horizon = len(test)
predictions = results.forecast(steps=forecast_horizon)

# Predict GDP for 2023
forecast_2023 = results.forecast(steps=1)[0]

# Plot the actual, predicted, and test GDP
years_train = df_SriLanka['Year'][:len(train)]
years_test = df_SriLanka['Year'][len(train):len(train) + len(test)]
years_all = df_SriLanka['Year']

plt.figure(figsize=(10, 6))
plt.plot(years_all, gdp_data, label='Actual GDP', color='blue')
plt.plot(years_test, test, label='Test Data', linestyle='--', color='red')
plt.plot(years_all[len(train):len(train) + len(predictions)], predictions, label='Predicted GDP', color='green')
plt.scatter(years_test, test, color='red', marker='o')  # Add circle markers for test data
plt.scatter(years_all.iloc[-1], forecast_2023, color='red')  # Mark the predicted value for 2023
plt.title('Sri Lanka GDP: Actual vs. Predicted')
plt.xlabel('Year')
plt.ylabel('GDP (USD)')
plt.legend()
plt.show()

print("Predicted GDP for 2023:", forecast_2023)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA

sns.set_style("darkgrid")

# Assuming you have already loaded the GDP data into df_SriLanka DataFrame

# Extract GDP data and years
gdp_data = df_SriLanka['GDP'].values
years = df_SriLanka['Year'].values

# Take first differences to make GDP values stationary
diff_gdp_data = np.diff(gdp_data)

# Plot stationary GDP values
plt.figure(figsize=(10, 6))
plt.plot(years[1:], diff_gdp_data, label='Stationary GDP', color='blue', linestyle='-')
plt.title('Sri Lanka GDP: Stationary Data', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('GDP Difference (USD)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='-', alpha=0.7)
plt.legend(loc='upper left', fontsize='large')
sns.despine()
plt.show()

# Fit ARIMA model on stationary data
model_stationary = ARIMA(diff_gdp_data, order=(1, 0, 1))
results_stationary = model_stationary.fit()

# Forecast on stationary data
forecast_horizon = 10
predictions_stationary = results_stationary.forecast(steps=forecast_horizon)

# Plot ARIMA predictions on stationary data
plt.figure(figsize=(10, 6))
plt.plot(years, gdp_data, label='Actual GDP', color='blue', linestyle='-')

# Generate predicted years
predicted_years = np.arange(years[-1] + 1, years[-1] + forecast_horizon + 1)

# Adjust length of predictions to match predicted_years
adjusted_predictions = np.append(gdp_data[-1], np.cumsum(predictions_stationary))[:forecast_horizon]

plt.plot(predicted_years, adjusted_predictions, label='Predicted GDP', color='orange', linestyle='-', marker='o')
plt.scatter(predicted_years[0], adjusted_predictions[0], s=200, color='red', label='Predicted GDP for 2023')
plt.title('Sri Lanka GDP: Actual vs. Predicted (Stationary Data)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('GDP (USD)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='-', alpha=0.7)
plt.legend(loc='upper left', fontsize='large')
sns.despine()
plt.show()

```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Invert differencing to get predicted GDP values
predicted_gdp = np.cumsum(predictions_stationary)

# Calculate MAE
mae = mean_absolute_error(gdp_data[-forecast_horizon:], predicted_gdp)
print("Mean Absolute Error (MAE):", mae)

# Calculate MSE
mse = mean_squared_error(gdp_data[-forecast_horizon:], predicted_gdp)
print("Mean Squared Error (MSE):", mse)

# Calculate RMSE
rmse = np.sqrt(mse)
print("Root Mean Squared Error (RMSE):", rmse)
```
